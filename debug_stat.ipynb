```python
from matplotlib import pyplot as plt
import os
import numpy as np
```

```python
base_dir = '/Users/yz9244/Downloads/debug'
qread_array = []
for i in range(20,21):
    qread = np.load(os.path.join(base_dir, 'qread_%03d.npy'%(i)))
    qread_array.append(qread[0].reshape(256, -1))
qread = np.concatenate(qread_array, axis=1)
print(qread.shape)
print('mean')
print(np.mean(qread, axis=1))
print('std')
print(np.std(qread, axis=1))
```

```python
plt.hist(qread[-1,:], bins='auto')
```
